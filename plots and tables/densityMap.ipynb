Code for creating a density map of the optimal transport plan 
and the marginals of the plan. <br />This code was used to create the
figures in section 5.1 of the thesis contained in https://github.com/Detaille/Semismooth-Newton

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

# read files
# read transportplan
plan = pd.read_csv('test.csv', header=None)

# read marginals
my = pd.read_csv('margBiNorm500.csv', header=None)
ny = pd.read_csv('margNorm500.csv', header=None)
```

```python
# plot the density map of the plan
p = plt.pcolormesh(plan)

# plot image without axis
plt.gca().set_axis_off()
plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, hspace = 0, wspace = 0)
plt.margins(0,0)
#plt.savefig("test.png")
```

```python
# plot marginal my
plt.plot(my.iloc[0], my.iloc[1], linewidth=2)
# save image without borders
plt.gca().set_axis_off()
plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, hspace = 0, wspace = 0)
plt.margins(0,0.01)
plt.xlim([0, 1])
plt.show()
#plt.savefig("margA500.png")
```

```python
# plot marginal ny
plt.plot(ny.iloc[0], ny.iloc[1], color="red", linewidth=2)
# save image without borders
plt.gca().set_axis_off()
plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, 
            hspace = 0, wspace = 0)
plt.margins(0,0.01)
#plt.savefig("margNorm500.png")
```
